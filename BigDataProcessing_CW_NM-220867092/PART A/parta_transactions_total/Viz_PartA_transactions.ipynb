```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df1 = pd.read_csv('transactions_total1.csv')
df1.columns = ['Date', 'Value']
df1.head()
```

```python
plt.figure(figsize = (20,10))
plt.xticks(rotation=90)
plt.xlabel('Month/Year', fontsize=12)
plt.ylabel('Total Transactions each month', fontsize=12)
x=df1['Date']
y=df1['Value']
plt.bar(x,y,color='navy')
```
