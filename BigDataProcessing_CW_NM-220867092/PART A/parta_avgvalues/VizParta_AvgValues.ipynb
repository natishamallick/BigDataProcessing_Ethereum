```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('new_average_values.csv')
df.columns = ['Date', 'Value']
df.head()
```

```python
plt.figure(figsize = (20,10))
plt.xticks(rotation=90)
plt.xlabel('Month/Year', fontsize=12)
plt.ylabel('Average Transactions each month', fontsize=12)
x=df['Date']
y=df['Value']
plt.bar(x,y,color='navy')
```
